# Featur Engineering 

```python
# Convert Date column to datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
# Extract date-related features
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['Weekday'] = df['Date'].dt.weekday
```

```python
# Profit margin per transaction (assuming revenue - cost, if cost is available)
df['Total_Profit'] = df['Total Amount Including Taxes'] - df['Total Amount Excluding Taxes']

# Tax burden as a percentage
df['Tax_Percentage'] = (df['Total Sales Tax @ 18%'] / df['Total Amount Excluding Taxes']) * 100
```

```python
# Count of transactions per customer and seller (indicating customer loyalty & seller activity)
df['Customer_Transaction_Count'] = df.groupby('Customer_NTN')['Customer_NTN'].transform('count')
df['Seller_Transaction_Count'] = df.groupby('Seller_NTN')['Seller_NTN'].transform('count')
```

```python
# Encoding categorical features (e.g., Business type, Seller & Customer Names)
df['Customer_Business_Encoded'] = df['Customer_Business'].astype('category').cat.codes
df['Seller_Name_Encoded'] = df['Seller_Name'].astype('category').cat.codes
df['Customer_Name_Encoded'] = df['Customer_NAME'].astype('category').cat.codes
```

```python
# Drop unnecessary text columns for modeling
columns_to_drop = ['Seller_NTN', 'Seller_Name', 'Seller_S.T.Reg.No', 'Seller_Address', 
                   'Customer_NTN', 'Customer_NAME', 'Customer_Tax_ID', 'Customer_Address', 
                   'Customer_Business', 'Date']
df_cleaned = df.drop(columns=columns_to_drop)
```
